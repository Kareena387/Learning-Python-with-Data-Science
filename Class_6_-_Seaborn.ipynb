```python
import seaborn as sns  
import matplotlib.pyplot as plt
```

```python
data = sns.load_dataset('iris')
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)

# setting the title using Matplotlib 
plt.title('Title using Matplotlib Function') 
  
plt.show()
```

```python
sns.lineplot(x="sepal_length", y="sepal_width", data=data)
sns.set_style('white') # dark, white, ticks
```

```python
tips = sns.load_dataset('tips')
tips
```

```python
sns.distplot(tips['total_bill'])
```

```python
sns.jointplot(x='total_bill', y='tip', data=tips, kind='scatter') # hex, reg
```

```python
sns.pairplot(tips) 
```

```python
sns.pairplot(tips, hue='sex')  #palette
```

### CATEGORICAL PLOTS

```python
sns.barplot(x='sex', y='total_bill', data=tips)
```

```python
sns.boxplot(x='day', y='total_bill', data=tips) #hue
```

```python
sns.violinplot(x='day', y='total_bill', data=tips) #hue
```

#### MATRIX PLOT

```python
tips.corr()
```
